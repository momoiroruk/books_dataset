```python
#Cohen effect size function, 
#which is the difference in means expressed in number of 
#standard deviations
def cohen_effect_size(column_1, column_2):
    
    t_mean = column_1.mean() - column_2.mean()
    var_1 = column_1.var()
    var_2 = column_2.var()
    n_1, n_2 = len(column_1), len(column_2)
    t_var = (n_1*var_1+n_2*var_2)/(n_1+n_2)
    t = t_mean/np.sqrt(t_var)
    return t
   
```

```python
#using cohen effect function
cohen_effect_size(data.ratings, data.text_reviews)
```
